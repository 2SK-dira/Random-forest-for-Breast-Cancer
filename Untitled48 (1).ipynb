```python
# -*- coding: utf-8 -*-
# TP Random Forest - Breast Cancer Wisconsin (Diagnostic) - UCI direct
# Auteur: Koudira
# Objectif: Implémentation conforme au cahier des charges + EDA approfondie + prévention des fuites
# Remarques intégrées: encodage catégoriel, déséquilibre des classes, choix du scaler (Standard/MinMax), EDA complète

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, classification_report, ConfusionMatrixDisplay,
    RocCurveDisplay, PrecisionRecallDisplay
)
from sklearn.tree import plot_tree
from sklearn.inspection import permutation_importance

# Optionnel: SMOTE pour illustrer la gestion de classes déséquilibrées
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
```

```python

# =========================
# Paramètres généraux
# =========================
RANDOM_STATE = 42
OUT = Path("outputs"); OUT.mkdir(exist_ok=True)

# Choix du scaler:
# 'standard' => StandardScaler (moyenne = 0, écart-type = 1), robuste pour comparaison inter-modèles, moins sensible aux outliers
# 'minmax'   => MinMaxScaler (mise à l'échelle dans [0,1]), utile pour algos basés sur distances, plus sensible aux outliers
SCALER_CHOICE = 'standard'  # changer en 'minmax' si souhaité

def make_scaler(choice):
    if choice == 'standard':
        return StandardScaler()
    elif choice == 'minmax':
        return MinMaxScaler()
    else:
        raise ValueError("SCALER_CHOICE doit être 'standard' ou 'minmax'.")
```

```python
# =========================
# 1) Chargement direct depuis UCI (sans téléchargement manuel)
# =========================
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
df = pd.read_csv(url, header=None)
df.head()
```

```python
# Définition des colonnes (UCI)
columns = ["ID", "Diagnosis"] + [f"feature_{i}" for i in range(1, 31)]
df.columns = columns

# Encodage de la variable catégorielle Diagnosis: M=1 (malin), B=0 (bénin)
df["Diagnosis"] = df["Diagnosis"].map({"M": 1, "B": 0})

# Séparation X/y et suppression de l'ID (non informatif)
X = df.drop(columns=["ID", "Diagnosis"])
y = df["Diagnosis"]
df.head()
```

```python
# =========================
# 2) EDA (Analyse exploratoire des données)
# =========================
print(f"Dimensions du dataset: {df.shape} (569 lignes, 32 colonnes dont 30 features)")
print("Aperçu des premières lignes:")
print(df.head())
```

```python
# 2.1 Répartition des classes (équilibre/déséquilibre)
class_counts = y.value_counts()
class_pct = y.value_counts(normalize=True).round(3)
print("\nRépartition des classes (counts):")
print(class_counts)
print("\nRépartition des classes (pourcentages):")
print(class_pct)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.title("Répartition des classes (0 = Bénin, 1 = Malin)")
plt.xlabel("Classe")
plt.ylabel("Nombre d'observations")
plt.tight_layout()
plt.savefig(OUT/"class_distribution.png")
plt.show()
```

```python
# 2.2 Statistiques descriptives
desc = X.describe()
print("\nStatistiques descriptives (features):")
print(desc)
desc.to_csv(OUT/"descriptive_stats.csv")
```

```python




# =========================
# 3) Split train/test 80/20 (stratifié)
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=RANDOM_STATE
)

# =========================
# 4) Pipeline RF avec scaler (prévention des fuites via Pipeline)
# =========================
scaler = make_scaler(SCALER_CHOICE)
rf_base = Pipeline([
    ("scaler", scaler),  # scaling appris sur les folds d'entraînement uniquement
    ("rf", RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1))
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

# Évaluation baseline en CV (train seulement)
scores_acc = cross_val_score(rf_base, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1)
scores_f1  = cross_val_score(rf_base, X_train, y_train, cv=cv, scoring="f1", n_jobs=-1)
scores_auc = cross_val_score(rf_base, X_train, y_train, cv=cv, scoring="roc_auc", n_jobs=-1)

print(f"\nBaseline CV Accuracy: {scores_acc.mean():.3f} ± {scores_acc.std():.3f}")
print(f"Baseline CV F1     : {scores_f1 .mean():.3f} ± {scores_f1 .std():.3f}")
print(f"Baseline CV ROC-AUC: {scores_auc.mean():.3f} ± {scores_auc.std():.3f}")
pd.DataFrame({
    "accuracy": scores_acc,
    "f1": scores_f1,
    "roc_auc": scores_auc
}).to_csv(OUT/"cv_baseline_scores.csv", index=False)
```

```python
# =========================
# 5) Hyperparameter tuning (plages conformes au TP)
# =========================
param_grid = {
    "rf__n_estimators": [100, 300, 500, 800, 1000],
    "rf__max_depth": [5, 10, 15, 20, 30],
    "rf__min_samples_split": [2, 5, 10, 15, 20],
    "rf__min_samples_leaf": [1, 2, 4, 6, 10],
    "rf__max_features": ["sqrt", "log2"]  # 'auto' est obsolète; 'sqrt' est l'équivalent conseillé
}

grid = GridSearchCV(
    estimator=rf_base,
    param_grid=param_grid,
    cv=cv,
    scoring="f1",  # compromis clinique (précision/rappel)
    n_jobs=-1,
    verbose=1
)
grid.fit(X_train, y_train)

print("\nMeilleurs hyperparamètres:", grid.best_params_)
print(f"Meilleur score CV (F1): {grid.best_score_:.3f}")
pd.Series(grid.best_params_).to_json(OUT/"best_params.json")

best_model = grid.best_estimator_
```

```python
# =========================
# 6) Évaluation finale sur le test set (non vu pendant tuning)
# =========================
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

def report_metrics(y_true, y_pred, y_proba):
    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred),
        "recall": recall_score(y_true, y_pred),
        "f1": f1_score(y_true, y_pred),
        "roc_auc": roc_auc_score(y_true, y_proba),
    }

metrics_test = report_metrics(y_test, y_pred, y_proba)
print("\nMetrics test (RF optimisé):", {k: round(v, 3) for k, v in metrics_test.items()})
print("\nClassification report:\n", classification_report(y_test, y_pred))
pd.Series(metrics_test).to_json(OUT/"metrics_rf_test.json")

# Visualisations test
fig_cm, ax_cm = plt.subplots(figsize=(5, 5))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax_cm, cmap="Blues")
plt.title("Matrice de confusion (RF optimisé)")
plt.tight_layout(); fig_cm.savefig(OUT/"confusion_matrix_rf.png"); plt.show()

fig_roc, ax_roc = plt.subplots(figsize=(5, 5))
RocCurveDisplay.from_predictions(y_test, y_proba, ax=ax_roc)
plt.title("Courbe ROC (RF optimisé)")
plt.tight_layout(); fig_roc.savefig(OUT/"roc_rf.png"); plt.show()

fig_pr, ax_pr = plt.subplots(figsize=(5, 5))
PrecisionRecallDisplay.from_predictions(y_test, y_proba, ax=ax_pr)
plt.title("Courbe Precision-Recall (RF optimisé)")
plt.tight_layout(); fig_pr.savefig(OUT/"pr_rf.png"); plt.show()
```

```python


# =========================
# 7) Visualisation d’un arbre du Random Forest
# =========================
rf_fitted = best_model.named_steps["rf"]
plt.figure(figsize=(22, 14))
plot_tree(
    rf_fitted.estimators_[0],
    feature_names=X.columns,
    class_names=["benign", "malignant"],
    filled=True,
    max_depth=3  # profondeur limitée pour lisibilité
)
plt.title("Random Forest - Visualisation d'un arbre (profondeur=3)")
plt.tight_layout(); plt.savefig(OUT/"rf_single_tree.png"); plt.show()

```

```python
# =========================
# 8) Amélioration 1: Sélection de features (Permutation Importance top-k)
# =========================
pi = permutation_importance(best_model, X_test, y_test, n_repeats=20,
                            random_state=RANDOM_STATE, n_jobs=-1)
imp = pd.DataFrame({"feature": X.columns, "importance": pi.importances_mean}) \
         .sort_values("importance", ascending=False)
imp.to_csv(OUT/"permutation_importance.csv", index=False)

top_k = 15
selected_features = imp["feature"].head(top_k).tolist()

plt.figure(figsize=(9, 7))
sns.barplot(x="importance", y="feature", data=imp.head(top_k), orient="h")
plt.title(f"Top {top_k} features par permutation importance")
plt.tight_layout(); plt.savefig(OUT/"permutation_importance_top15.png"); plt.show()

# Refit RF sur sous-espace sélectionné (mêmes hyperparamètres optimisés)
X_train_sel = X_train[selected_features]
X_test_sel  = X_test[selected_features]

best_model_sel = Pipeline([
    ("scaler", make_scaler(SCALER_CHOICE)),
    ("rf", RandomForestClassifier(
        random_state=RANDOM_STATE, n_jobs=-1,
        n_estimators=grid.best_params_["rf__n_estimators"],
        max_depth=grid.best_params_["rf__max_depth"],
        min_samples_split=grid.best_params_["rf__min_samples_split"],
        min_samples_leaf=grid.best_params_["rf__min_samples_leaf"],
        max_features=grid.best_params_["rf__max_features"],
    ))
])
best_model_sel.fit(X_train_sel, y_train)
y_pred_sel = best_model_sel.predict(X_test_sel)
y_proba_sel = best_model_sel.predict_proba(X_test_sel)[:, 1]
metrics_sel = report_metrics(y_test, y_pred_sel, y_proba_sel)
print("\nMetrics test (RF avec sélection de features):", {k: round(v, 3) for k, v in metrics_sel.items()})
pd.Series(metrics_sel).to_json(OUT/"metrics_rf_selected_features.json")
```

```python
# =========================
# 9) Amélioration 2: Ensemble avec Gradient Boosting (comparaison)
# =========================
gb = Pipeline([
    ("scaler", make_scaler(SCALER_CHOICE)),
    ("gb", GradientBoostingClassifier(random_state=RANDOM_STATE))
])
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
# GradientBoostingClassifier ne fournit predict_proba si loss='log_loss' (par défaut),
# sinon fallback sur decision_function normalisé
try:
    y_proba_gb = gb.predict_proba(X_test)[:, 1]
except Exception:
    # Normalisation simple pour PR/ROC si nécessaire
    y_scores = gb.decision_function(X_test)
    y_proba_gb = (y_scores - y_scores.min()) / (y_scores.max() - y_scores.min() + 1e-9)

metrics_gb = report_metrics(y_test, y_pred_gb, y_proba_gb)
print("\nMetrics test (Gradient Boosting):", {k: round(v, 3) for k, v in metrics_gb.items()})
pd.Series(metrics_gb).to_json(OUT/"metrics_gb_test.json")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_base = Pipeline([
    ("scaler", make_scaler(SCALER_CHOICE)),
    ("gb", GradientBoostingClassifier(random_state=RANDOM_STATE))
])

param_grid_gb = {
    "gb__n_estimators": [100, 200, 300, 500],
    "gb__learning_rate": [0.03, 0.05, 0.1],
    "gb__max_depth": [2, 3, 4],
    "gb__subsample": [1.0, 0.9, 0.8],
    "gb__min_samples_split": [2, 5, 10],
    "gb__min_samples_leaf": [1, 2, 4],
    "gb__max_features": [None, "sqrt", "log2"]
}

grid_gb = GridSearchCV(
    estimator=gb_base,
    param_grid=param_grid_gb,
    cv=cv,
    scoring="f1",
    n_jobs=-1,
    verbose=1
)
grid_gb.fit(X_train, y_train)

print("\n[GB] Best hyperparameters:", grid_gb.best_params_)
print(f"[GB] Best CV F1: {grid_gb.best_score_:.3f}")
pd.Series(grid_gb.best_params_).to_json(OUT/"best_params_gb.json")

best_gb = grid_gb.best_estimator_
y_pred_gb = best_gb.predict(X_test)
y_proba_gb = best_gb.predict_proba(X_test)[:, 1]

metrics_gb = report_metrics(y_test, y_pred_gb, y_proba_gb)
pd.Series(metrics_gb).to_json(OUT/"metrics_gb_test.json")
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_base = Pipeline([
    ("scaler", make_scaler(SCALER_CHOICE)),
    ("ada", AdaBoostClassifier(random_state=RANDOM_STATE))
])

param_grid_ada = {
    "ada__n_estimators": [100, 200, 300],
    "ada__learning_rate": [0.5, 1.0, 1.5],
    # Base estimator depth indirect: AdaBoost par défaut utilise des stumps.
    # Pour plus de contrôle, on peut utiliser des DecisionTreeClassifier max_depth=2/3.
}

grid_ada = GridSearchCV(
    estimator=ada_base,
    param_grid=param_grid_ada,
    cv=cv,
    scoring="f1",
    n_jobs=-1,
    verbose=1
)
grid_ada.fit(X_train, y_train)

print("\n[Ada] Best hyperparameters:", grid_ada.best_params_)
print(f"[Ada] Best CV F1: {grid_ada.best_score_:.3f}")
pd.Series(grid_ada.best_params_).to_json(OUT/"best_params_ada.json")

best_ada = grid_ada.best_estimator_
y_pred_ada = best_ada.predict(X_test)

# AdaBoost fournit predict_proba
y_proba_ada = best_ada.predict_proba(X_test)[:, 1]
metrics_ada = report_metrics(y_test, y_pred_ada, y_proba_ada)
pd.Series(metrics_ada).to_json(OUT/"metrics_ada_test.json")
```

```python
from xgboost import XGBClassifier

xgb_base = XGBClassifier(
    random_state=RANDOM_STATE,
    eval_metric="logloss",
    use_label_encoder=False
)

param_grid_xgb = {
    "n_estimators": [200, 400, 600],
    "learning_rate": [0.03, 0.05, 0.1],
    "max_depth": [3, 4, 5],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
    "gamma": [0, 0.1]
}

grid_xgb = GridSearchCV(
    estimator=xgb_base,
    param_grid=param_grid_xgb,
    cv=cv,
    scoring="f1",
    n_jobs=-1,
    verbose=1
)
grid_xgb.fit(X_train, y_train)

print("\n[XGB] Best hyperparameters:", grid_xgb.best_params_)
print(f"[XGB] Best CV F1: {grid_xgb.best_score_:.3f}")
pd.Series(grid_xgb.best_params_).to_json(OUT/"best_params_xgb.json")

best_xgb = grid_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)
y_proba_xgb = best_xgb.predict_proba(X_test)[:, 1]

metrics_xgb = report_metrics(y_test, y_pred_xgb, y_proba_xgb)
pd.Series(metrics_xgb).to_json(OUT/"metrics_xgb_test.json")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier

# Utiliser le meilleur RF (grid.best_estimator_) et le meilleur GB (grid_gb.best_estimator_)
estimators = [
    ("rf", best_model),   # RF optimisé via GridSearch initial
    ("gb", best_gb)       # GB optimisé
]

stack = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    stack_method="predict_proba",
    passthrough=False,    # si True: ajoute X original au méta-apprenant
    n_jobs=-1
)
stack.fit(X_train, y_train)

y_pred_stack = stack.predict(X_test)
# Pour ROC-AUC, on a besoin des probabilités de la classe 1
y_proba_stack = stack.predict_proba(X_test)[:, 1]

metrics_stack = report_metrics(y_test, y_pred_stack, y_proba_stack)
print("\n[Stack RF+GB] Metrics test:", {k: round(v, 3) for k, v in metrics_stack.items()})
pd.Series(metrics_stack).to_json(OUT/"metrics_stack_rf_gb_test.json")
```

```python
estimators_ext = [
    ("rf", best_model),
    ("gb", best_gb),
    ("ada", best_ada),
    ("xgb", best_xgb)
]
stack_ext = StackingClassifier(
    estimators=estimators_ext,
    final_estimator=LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    stack_method="predict_proba",
    n_jobs=-1
)
stack_ext.fit(X_train, y_train)
y_pred_stack_ext = stack_ext.predict(X_test)
y_proba_stack_ext = stack_ext.predict_proba(X_test)[:, 1]
metrics_stack_ext = report_metrics(y_test, y_pred_stack_ext, y_proba_stack_ext)
pd.Series(metrics_stack_ext).to_json(OUT/"metrics_stack_all_test.json")
```
